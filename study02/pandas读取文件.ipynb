```python
#数据清洗：是指发现并纠正数据中可识别的错误的过程，包括处理缺失值，重复值，异常值，统一数据格式，保证数据的一致性

pd.set_option('display.min_rows',30) #设置显示行数，表现为前面一半，后面一半，中间折叠（默认十行）

#1.读取数据
df=pd.read_csv('csv_data/02.csv')

#2.数据清洗
#2.1缺失值处理
df.isnull() #查看缺失值

#2.1.1删除缺失值
df.dropna() #删除缺失值所在行
df.dropna(axis=1) #axis值设为1，表示删除所在列，默认为0默认删除行
df

#2.1.2填充缺失值
df.fillna('---') #填充缺失值
df.ffill() #填充缺失值，填充上一行的数据
df.bfill() #填充缺失值，填充下一行的数据
```

```python
#2.2 重复值处理
#2.2.1 查看重复值
df.duplicated() #查看重复值（所有列的数据都一样）
df.duplicated(subset=['工作']) #subset指定列的数据重复

#2.2.2 删除重复值-->drop_duplicates
df.drop_duplicates(subset=['工作'],keep='first') #keep：保留重复值的第一行或者最后一行，False表示为全部删除
```

```python
#2.3 异常值的处理
#2.3.1查看异常值
df[df['年龄']<18]

#2.3.2 删除异常值
df.drop(df[df['年龄']<18].index)

#2.3.3 修复异常值
df['年龄']=df['年龄'].abs() #绝对值
df
```

```python
#2.4 数据格式处理
df['日期']=df['日期'].str.replace('/','-')
df
```

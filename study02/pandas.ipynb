```python
#pandas：是一个功能强大的结构化数据分析的工具集，底层是基于Numpy构建的，无论是在数据分析领域，还是
# 大数据开发场景中都有显著的优势
#核心：DataFrame(类似表格)，Series(类似表格中的一列)
import pandas as pd

df1=pd.DataFrame([
    {'姓名':'张三','性别':'男','年龄':16,'语文':78},
    {'姓名':'李四','性别':'男','年龄':17,'语文':88}
])

print(f"最高分:{df1['语文'].max()}，最低分{df1['语文'].min()}，平均分:{df1['语文'].mean()}")
```

```python
#DataFrame的构建方式：

#df.index：索引行数
#df.columns：表头数据
#df.values：表格数据
#df.dtypes：每一列的类型
#df.size：获取单元格的数量
#df.shape：获取数据维度（行，列）

df1=pd.DataFrame([
    {'姓名':'张三','性别':'男','年龄':16,'语文':78},
    {'姓名':'李四','性别':'男','年龄':17,'语文':88}
])

df2=pd.DataFrame({
    '姓名':['张三','李四'],
    '语文':[77,89],
    '性别':['男','男']
})

df3=pd.DataFrame([
    ('张三','男',78),
    ('李四','男',55)
],columns=['姓名','性别','年龄'])

df4=pd.DataFrame([
    ['张三','男',78],
    ['李四','男',64]
],columns=['姓名','性别','成绩'],index=['a','b'])

df4
```

```python
#Matplotlib：是一个功能强大的数据可视化开源Python库，也是Python中使用最多的图形绘图库，可以创建
# 静态，动态，交互式的图表

import matplotlib.pyplot as plt
from matplotlib.lines import lineStyles
from mpl_toolkits.axisartist import Axes
from pydantic import color

x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,4,5,6,7,8,9]

#绘制折线图
plt.plot(x,y)
#展示图表
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,4,5,6,7,8,9]

#展示中文
plt.rcParams['font.sans-serif']=['SimHei'] #设置中文字体为黑体

plt.figure(figsize=(5,5)) #设置画布大小，宽，高
#绘制折线图 --> 如果没有画布，会自动创建画布
plt.plot(x,y,label='北京')

#设置折线图的详细信息
plt.title('气温变化折线图',fontsize=15) #fontsize：设置字体大小

plt.xlabel('时间') #x轴标签
plt.ylabel('温度') #y轴标签

plt.xticks(x[::1]) #从哪里开始展示，到哪里，一步多少
plt.yticks(y[::1])

plt.grid(linestyle='--') #显示网格

#loc：表示要展示的位置
plt.legend(loc='best') #显示图例

#展示图表
plt.show()
```

```python
#创建子图：为了能同时展示多个图表，便于图表之间数据的直观对比和分析，更高效，更专业地组织和呈现复杂的可视化信息，通常会在一个画布上创建多个子图

#figure：画布对象  axes：子图
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,6),dpi=100) #nrows：行  ncols：列 figsize：设置画布大小  dpi：设置分辨率

#图一：柱状图（世界石油储备）--> bar()
countries=['中国','美国','印度','加拿大','伊朗']
values=[35,23,18,21,56]

axes1:Axes=axes[0]
axes1.bar(countries,values,width=0.6,color='blue') #width:柱子宽度 color：颜色
axes1.set_title('世界石油储备',fontsize=15)
axes1.set_xlabel('国家')
axes1.set_ylabel('石油储备')

axes1.grid(linestyle='--',alpha=0.3) #网格线 alpha：透明度

#图二：饼状图（世界人口）--->擅长比例分析
countries2=['印度','中国','美国','印尼']
values=[14.51,14.09,3.4,2.83]

axes2:Axes=axes[1]
#饼状图第一个要传值
axes2.pie(values,labels=countries2,autopct='%1.1f%%') #autopct：展示比率,前后保留几位
axes2.set_title('世界人口比例',fontsize=15)
axes2.legend(loc='best')

#保存图片
plt.savefig('csv_data/test.png')
plt.show()
```

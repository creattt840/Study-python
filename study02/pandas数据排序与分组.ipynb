```python
import pandas as pd

#1.读取数据
df=pd.read_csv('csv_data/02.csv',nrows=10)  #nrows：只读取前n行

#2.排序
#2.1 根据 年龄 倒序排序
df.sort_values('年龄',ascending=False) #ascending默认为True升序排序

#2.2 根据 年龄 升序排序
df.sort_values('年龄',ascending=True)

#2.3 根据年龄升序排序，年龄一样，根据时间倒序排序
df.sort_values(['年龄','日期'],ascending=[True,False])
```

```python
#分组
#分组后求和，()后面的[]是根据什么进行计算
df.groupby('产品类别')['销售额'].sum()
#类似的还有 .count() .max() .min() .mean()

#agg一次性执行多项统计
df.groupby('产品类别')['单价'].agg(['mean','max','min'])

df.groupby('产品类别').agg({'销售数量':'sum','销售金额':'sum','单价':'mean'})
```

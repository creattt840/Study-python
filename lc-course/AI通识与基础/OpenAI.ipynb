OpenAI调用大模型


初始化客户端

```python
from openai import OpenAI
from dotenv import load_dotenv
import os

#先加载环境变量
load_dotenv()

#再读取环境变量
api_key=os.getenv("DEEPSEEK_APIKEY")

#创建OpenAI 客户端
client=OpenAI(
    #从环境变量中获取APIKEY使用到python-dotenv
    api_key=api_key,
    base_url="https://api.deepseek.com"
)
```

访问模型

```python
print("正在调用大模型")
response =client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role":"system","content":"你是一个友好的AI助教。"},
        {"role":"user","content":"你好，你是谁？"}
    ],
    temperature=0.5,
    stream=False
)
print("调用大模型成果....")
```

获取模型响应结果

```python
print(response.choices[0].message.content)
```

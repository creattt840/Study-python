基于名称判断模型提供商

使用init_chat_model函数，你需要从LangChain支持的模型提供商中选择一个模型，而LangChain根据模型名称自动初始化与模型的连接，非常方便

```python
#导入LangChain的出初始化建模的函数
from langchain.chat_models import init_chat_model
from langchain_core.messages.block_translators import openai

#调用init_chat_model函数初始化模型
#参数model用来指定模型名称，LangChain会根据模型名字自动设定base_url，并从环境变量中获取api_key
model=init_chat_model(model="deepseek-chat")
```

```python
#init_chat_model返回的模型会根据模型名称自动确定其类型
print(type(model))
```

如果我们要访问的模型，不是LangChain支持的模型，我们必须自定义模型参数来访问。                        我们需要在环境变量中定义base_url和api_key                                                 然后在init_chat_model中指定model,model_provider,base_url和api_key

```python
#我们收到加载在环境变量中的base_url和api_key
import os

base_url=os.getenv("BASE_URL")
api_key=os.getenv("API_KEY")

model=init_chat_model(
    model="qwen-max", #模型名称
    model_provider=openai, #如果是LangChain不支持的模型需要指定模型提供者，阿里兼容openai，所以这里用openai
    base_url=base_url,
    api_key=api_key
)
```

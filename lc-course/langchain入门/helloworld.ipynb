1.加载环境变量

```python
#1.加载环境变量
from dotenv import load_dotenv

load_dotenv()
```

2.定义工具

```python
#2.定义工具
from langchain.tools import tool

@tool
def getWeather(location:str)->str:
    """
    查询天气的工具
    :param location: 地址
    :return: 返回当地的天气
    """
    return f"在{location}的天气是晴天"
```

3.创建Agent

```python
#3.创建Agent
from langchain.agents import create_agent

agent = create_agent(
    "deepseek-chat",
    tools=[getWeather]
)
```

4.发起调用

```python
#4.调用Agent
print("正在调用大模型...")
response = agent.invoke({
    "messages":[
        {"role":"user",
         "content":"杭州今天的天气怎么样"}
    ]
})
```
